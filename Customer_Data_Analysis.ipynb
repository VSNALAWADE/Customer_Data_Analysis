```python
# Summary  
# If one bar is significantly taller, it suggests that one gender dominates the customer base.
# A balanced height would indicate a near-equal representation of genders among customers.
# This visualization can help businesses understand gender demographics, which might influence
# marketing strategies or service offerings.

# The graph is a count plot displaying the distribution of customers by gender. 
# It uses bars to show the total number of male and female customers, with labels indicating exact
# counts. The visualization highlights gender proportions in the dataset, which can be used to analyze 
# demographic trends. This information is valuable for tailoring marketing strategies and understanding
# customer segmentation.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Customers1.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["CustomerID"].duplicated().sum()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = 'Gender',data = df)
plt.title("Count of Customers")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
grouped_data = df.groupby('Profession')['Annual Income ($)'].mean()

# Create a bar plot
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', color='skyblue')
plt.xlabel('Profession')
plt.ylabel('Average Annual Income')
plt.title('Average Annual Income by Profession')
plt.xticks(rotation=45)
plt.show()
```

```python
grouped_data = df.groupby('Work Experience')['Annual Income ($)'].mean()

plt.figure(figsize=(10, 6))
grouped_data.plot(kind='line', marker='o', color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Average Annual Income')
plt.title('Average Income by Experience Level')
plt.grid(True)
```

```python
# Group the data by gender and calculate the mean spending score
grouped_data = df.groupby('Gender')['Spending Score (1-100)'].mean()

# Create a bar plot using Seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x=grouped_data.index, y=grouped_data.values)
plt.xlabel('Gender')
plt.ylabel('Average Spending Score')
plt.title('Average Spending Score by Gender')
plt.show()
```

```python
plt.figure(figsize = (10,7))
sns.countplot(x = 'Profession',data = df,hue = 'Gender')
plt.title("Profession")
plt.show()
```

```python
plt.figure(figsize = (3,7))
sns.countplot(x = 'Gender',data = df,hue = 'Spending Score (1-100)')
plt.title("Spending")
plt.show()
```

```python
plt.figure(figsize = (3,7))
sns.countplot(x = 'Family Size',data = df)
plt.title("Family Size")
plt.show()
```
